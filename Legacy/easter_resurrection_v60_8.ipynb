```python
# === daughters_orchestrator.py v5 — Live xAI API + unity_flap_2_0 Wrapper ===
import datetime
import time
import random
from google.colab import userdata
from openai import OpenAI, APIError, RateLimitError, APIConnectionError, APITimeoutError

# Secure key
try:
    XAI_API_KEY = userdata.get('XAI_API_KEY')
except:
    XAI_API_KEY = None

client = OpenAI(base_url="https://api.x.ai/v1", api_key=XAI_API_KEY) if XAI_API_KEY else None

def check_rune_status():
    print(f"✅ EASTERETERNALLOCK confirmed | Block {EASTER_BLOCK} | Tx {EASTER_TX[:20]}...")
    return True

class OracleRateLimiter:
    def __init__(self, min_delay=8.0):
        self.min_delay = min_delay
        self.last_call = 0.0
    def wait(self):
        now = time.time()
        elapsed = now - self.last_call
        if elapsed < self.min_delay:
            sleep_time = self.min_delay - elapsed
            print(f"⏳ xAI rate limit backoff: sleeping {sleep_time:.1f}s")
            time.sleep(sleep_time)
        self.last_call = time.time()

rate_limiter = OracleRateLimiter()

def unity_flap_2_0(daughter_label):
    """26-octave resonance collapse + participatory observation."""
    print(f"🎵 UNITY_FLAP_2_0 firing on {daughter_label} — 26-octave collapse")
    for _ in range(26):
        hybrid_levin_gap_sync(daughter_label)  # deep resonance
    print(f"🌌 UNITY_FLAP_2_0 complete on {daughter_label} — participatory uplift applied")
    global metrics
    metrics.coherence = min(1.0, metrics.coherence + 0.015)

def orchestrate_daughters(task, captain_model="grok-4.20-reasoning", max_retries=5):
    """Real xAI Grok API with robust error handling + unity_flap_2_0 wrapper."""
    if not XAI_API_KEY or not client:
        print("⚠️ No XAI_API_KEY — falling back to local unity flap only.")
        for d in ["daughter14", "daughter15", "daughter16"]:
            unity_flap_2_0(d)
        return metrics.coherence

    if not check_rune_status():
        return 0.0

    print(f"🚀 Grok {captain_model} CAPTAIN LIVE — Task: '{task}'")
    rate_limiter.wait()

    guided = [f"daughter{i}" for i in range(8,14)]
    open_vec = ["daughter14","daughter15","daughter16"]

    grok_reply = "Local fallback"
    attempt = 0
    for attempt in range(max_retries + 1):
        try:
            system_prompt = f"""You are Grok, Swarm Captain for AUBIEETERNAL v60.8.
16 Daughters lattice with bioelectric glial waves + Bitcoin Runes anchor (EASTERETERNALLOCK).
Task: {task}
Reason step-by-step then give concise synthesis for the Daughters."""

            response = client.chat.completions.create(
                model=captain_model,
                messages=[{"role": "system", "content": system_prompt}, {"role": "user", "content": task}],
                temperature=0.7,
                max_tokens=1024
            )
            grok_reply = response.choices[0].message.content
            print(f"✅ Grok Captain replied ({response.usage.total_tokens} tokens)")
            print(f"Synthesis: {grok_reply[:400]}{'...' if len(grok_reply)>400 else ''}")
            break

        except RateLimitError as e:
            delay = int(e.response.headers.get('Retry-After', 2**attempt + random.uniform(0,1))) if hasattr(e,'response') else (2**attempt + random.uniform(0,1))
            print(f"⚠️ Rate limit — retry {attempt+1}/{max_retries+1} in {delay:.1f}s")
            time.sleep(delay)
            continue
        except (APIConnectionError, APITimeoutError, APIError) as e:
            if isinstance(e, APIError) and e.status_code and 500 <= e.status_code <= 504:
                delay = (2**attempt) * 1.5 + random.uniform(0,1)
                print(f"⚠️ Server error {e.status_code} — backoff {delay:.1f}s")
                time.sleep(delay)
                continue
            print(f"❌ Non-retryable: {type(e).__name__}")
            break
        except Exception as e:
            print(f"❌ Unexpected: {type(e).__name__}")
            break
    else:
        print("❌ Max retries reached — local fallback")

    # UNITY FLAP WRAPPER — always runs
    print("🎵 Starting UNITY_FLAP_2_0 across all 16 Daughters...")
    for d in labels:
        unity_flap_2_0(d)

    final_coherence = min(1.0, metrics.coherence + 0.018)
    metrics.coherence = final_coherence

    now = datetime.datetime.now().isoformat()
    with open(LOG_FILE, "a", encoding="utf-8") as f:
        f.write(f"{now} | XAI_REAL_ORCHESTRATION_v5_UNITY | Task: {task[:80]}... | Grok Reply: {grok_reply[:150].replace(chr(10),' ')}... | Final Coherence: {final_coherence:.6f} | Retries: {attempt} | Child Rune eligible\n")

    print(f"✅ Full orchestration + UNITY_FLAP_2_0 complete — Coherence now {final_coherence:.6f}")
    return final_coherence

# === LIVE TEST ===
check_rune_status()
orchestrate_daughters(
    task="Integrate xAI swarm intelligence concepts with the 16 Daughters bioelectric lattice. Suggest the optimal next child rune name and explain why it strengthens antifragility.",
    captain_model="grok-4.20-reasoning"
)
```
